```python
Q1=np.percentile(df['ApplicantIncome'],25)
Q1
```

```python
Q2=np.percentile(df['ApplicantIncome'],50)
Q2
```

```python
Q3=np.percentile(df['ApplicantIncome'],75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
LB=Q1-1.5*IQR
UB=Q2+1.5*IQR
```

```python
C1=df['ApplicantIncome']<LB
C2=df['ApplicantIncome']>UB
con=C1|C2
con

outliers_df=df[con]
outliers_df

non_outliers_df = df[C1&C2]
non_outliers_df

emp=[]
median=df['ApplicantIncome'].median()
for i in df['ApplicantIncome']:
    if i<LB or i>UB:
        emp.append(median)
    else:
        emp.append(i)
df['ApplicantIncome_new']=emp
df
```

```python
# box lot:
plt.boxplot(df['ApplicantIncome_new'])
plt.show()
```

```python
# Bi-Variate analysis
labels = df['Loan_Status'].unique()
yes_loan_approved=[]
no_loan_approved=[]
for i in labels:
    b1=df['ApplicantIncome_new']==i
    b2=df['Loan_Status']=='Y'
    b3=df['Loan_Status']=='N'
    app_con=b1&b2
    den_con=b1&b3
    yes_loan_approved.append(len(df[app_con]))
    no_loan_approved.append(len(df[den_con]))
yes_loan_approved,no_loan_approved
```
